```python
results = results.drop(['HTHG', 'HTAG', 'HTR', 'Referee'], axis=1)
```

```python
results = results.drop(['HF', 'AF', 'HY', 'AY', 'HR', 'AR'], axis=1)
```

```python
results = results.drop(['Season', 'DateTime'], axis=1)
```

```python
col_labels = results.columns
col_labels
```

```python
for c in col_labels:
    no_missing = results[c].isnull().sum()
    if no_missing > 0:
        print(c)
        print(no_missing)
    else:
        print(c)
        print("No missing values")
        print(' ')
```

```python
results['FTR']=results['FTR'].map({'H':1,'D':0, 'A':-1})
results.head()
```

```python
for c in col_labels:
    no_missing = results[c].isnull().sum()
    if no_missing > 0:
        print(c)
        print(no_missing)
    else:
        print(c)
        print("No missing values")
        print(' ')
```

```python
HomeTeams = results.HomeTeam.unique()
HomeTeams
```

```python
results["HomeTeam"].value_counts()
```

```python
HomeTeamsMapping = {}

count = 0
for team in HomeTeams:
    HomeTeamsMapping[team] = 1 + count
    count+=1
```

```python
HomeTeamsMapping
```

```python
results['HomeTeam'] = results['HomeTeam'].replace(HomeTeamsMapping)
```

```python
results['AwayTeam'] = results['AwayTeam'].replace(HomeTeamsMapping)
```
