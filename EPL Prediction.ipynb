```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

results = pd.read_csv (r"results.csv", encoding= 'unicode_escape')
```

```python
col_labels = results.columns
col_labels
```

```python
for c in col_labels:
    no_missing = results[c].isnull().sum()
    if no_missing > 0:
        print(c)
        print(no_missing)
    else:
        print(c)
        print("No missing values")
        print(' ')
```

```python
results = results.dropna()
```

```python
results['FTR']=results['FTR'].map({'H':1,'D':0, 'A':-1})
results['HTR']=results['HTR'].map({'H':1,'D':0, 'A':-1})
```

```python
fulltime = results.drop(['HTHG', 'HTAG', 'HTR', 'Referee', 'HS', 'AS', 'HST', 'AST', 'HC', 'AC',
       'HF', 'AF', 'HY', 'AY', 'HR', 'AR'], axis=1)

fulltime.head()
```
