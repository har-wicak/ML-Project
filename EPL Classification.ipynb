```python
results['FTR']=results['FTR'].map({'H':1,'D':0, 'A':-1})
```

```python
results['HTR']=results['HTR'].map({'H':1,'D':0, 'A':-1})
```

```python
HomeTeams = results.HomeTeam.unique()
HomeTeams
```

```python
results["HomeTeam"].value_counts()
```

```python
TeamsMapping = {}

count = 0
for team in HomeTeams:
    TeamsMapping[team] = 1 + count
    count+=1
```

```python
TeamsMapping
```

```python
results['HomeTeam'] = results['HomeTeam'].replace(TeamsMapping)
```

```python
results['AwayTeam'] = results['AwayTeam'].replace(TeamsMapping)
```

```python
results.head()
```

```python
results["HTHG"] = results["HTHG"].astype(int)
results["HTAG"] = results["HTAG"].astype(int)
results["HTR"] = results["HTR"].astype(int)
results["HS"] = results["HS"].astype(int)
results["AS"] = results["AS"].astype(int)
results["HST"] = results["HST"].astype(int)
results["AST"] = results["AST"].astype(int)
results["HC"] = results["HC"].astype(int)
results["AC"] = results["AC"].astype(int)
results
```

```python
# Full Time and Half Time Goal Difference (FTGD and HTGD respectively)
FTGD = results["FTHG"] - results["FTAG"]
results["FTGD"] = FTGD
HTGD = results["HTHG"] - results["HTAG"]
results["HTGD"] = HTGD
```

```python
# Halftime and Fulltime Winner
FTWinner = []
for i in range(len(results)):
    if results.iloc[i, results.columns.get_loc("FTGD")] > 0:
        FTWinner += [results.iloc[i, results.columns.get_loc("HomeTeam")]]
    elif results.iloc[i, results.columns.get_loc("FTGD")] < 0:
        FTWinner += [results.iloc[i, results.columns.get_loc("AwayTeam")]]
    else:
        FTWinner += [0]

results["FTWinner"] = FTWinner

HTWinner = []
for i in range(len(results)):
    if results.iloc[i, results.columns.get_loc("HTGD")] > 0:
        HTWinner += [results.iloc[i, results.columns.get_loc("HomeTeam")]]
    elif results.iloc[i, results.columns.get_loc("HTGD")] < 0:
        HTWinner += [results.iloc[i, results.columns.get_loc("AwayTeam")]]
    else:
        HTWinner += [0]

results["HTWinner"] = HTWinner
results.head()
```

```python
Comeback = []
for i in range(len(results)):
    if results.iloc[i, results.columns.get_loc("FTWinner")] != results.iloc[i, results.columns.get_loc("HTWinner")] and results.iloc[i, results.columns.get_loc("HTWinner")] != 0 and results.iloc[i, results.columns.get_loc("FTWinner")] != 0:
        Comeback += [1]
    else:
        Comeback += [0]

results["Comeback"] = Comeback
```

```python
results = results.drop(['Season', 'FTHG', 'FTAG', 'HTHG', 'HTAG',
       'HTR', 'FTGD', 'HTGD', 'FTWinner','HTWinner'], axis=1)
```

```python
X = results[['HomeTeam', 'AwayTeam', 'HS', 'AS', 'HST', 'AST', 'HC', 'AC',
       'Comeback']]

Y = results[['FTR']]
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3,random_state=1)
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
svm_classifier = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
svm_classifier.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = svm_classifier.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

# Model Precision: what percentage of positive tuples are labeled as such?
print("Precision:",metrics.precision_score(y_test, y_pred, average='micro'))

# Model Recall: what percentage of positive tuples are labelled as such?
print("Recall:",metrics.recall_score(y_test, y_pred, average='micro'))
```
